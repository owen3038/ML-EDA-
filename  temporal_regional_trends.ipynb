```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
df = pd.read_csv("cleaned_data_v1.csv")

# Display first few rows
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure 'Usage_Frequency' is categorical
df["Usage_Frequency"] = df["Usage_Frequency"].astype(str)  

plt.figure(figsize=(8,5))

sns.histplot(data=df, x="Age", bins=10, kde=True, hue="Usage_Frequency", palette="coolwarm")

plt.title("Age Distribution and Activity Level")
plt.xlabel("Age")
plt.ylabel("User Count")

# Explicitly pass labels to the legend
plt.legend(title="Usage Frequency", labels=df["Usage_Frequency"].unique())

plt.show()
```

```python
# Convert 'Date' column to datetime format (if it exists)
if "Date" in df.columns:
    df["Date"] = pd.to_datetime(df["Date"])

    plt.figure(figsize=(10,5))
    sns.lineplot(data=df, x="Date", y="Primary_App", hue="Primary_App", estimator="count")
    plt.title("Dating App Preference Over Time")
    plt.xlabel("Time")
    plt.ylabel("User Count")
    plt.xticks(rotation=45)
    plt.show()
else:
    print("No date column found in the dataset.")
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="Location", hue="Usage_Frequency", palette="viridis")
plt.xticks(rotation=45)
plt.title("Usage Frequency by Location")
plt.xlabel("Location")
plt.ylabel("User Count")
plt.legend(title="Usage Frequency")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="Location", y="Daily_Usage_Time", hue="Location", palette="coolwarm", legend=False)

plt.xticks(rotation=45)
plt.title("Daily Usage Time by Location")
plt.xlabel("Location")
plt.ylabel("Daily Usage (Minutes)")
plt.show()
```
